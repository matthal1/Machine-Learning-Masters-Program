```python
def image_show(X, title):
    img = plt.imshow(X)
    img.set_cmap('gray')
    plt.axis("off")
    plt.title(title)
    plt.show()
def rgb_gray(filename):
    A = imread(filename)
    return np.mean(A, -1)
```
